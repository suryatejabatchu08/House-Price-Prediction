```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return(float(temp[0]) + float(temp[1]))/2
    try :
        return float(x)
    except :
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convertRange)
```

```python
df = df.drop(columns=['size'])
```

```python
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
```

```python
df['location'] = df['location'].apply(lambda x: x.strip())
location_count = df['location'].value_counts()
```

```python
location_count_less_10 = location_count[location_count<=10]
location_count_less_10
```

```python
df['location'] = df['location'].apply(lambda x: 'other' if x in location_count_less_10 else x)
```

## Removing Outliers

```python
(df['total_sqft']/df['bhk']).describe()
```

```python
df = df[((df['total_sqft']/df['bhk']) >= 300)]
df.describe()
```

```python
df.shape
```

```python
df.price_per_sqft.describe()
```

```python
def remove_outliers_sqft(data):  
    df_output = pd.DataFrame()  
    for key, subdf in data.groupby('location'):  
        m = np.mean(subdf.price_per_sqft)  
        st = np.std(subdf.price_per_sqft)  

        gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]  
        df_output = pd.concat([df_output, gen_df], ignore_index=True)  

    return df_output
df = remove_outliers_sqft(df)
df.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean' : np.mean(bhk_df.price_per_sqft),
                
                'std' : np.std(bhk_df.price_per_sqft),
                
                'count' : bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
              exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
df = bhk_outlier_remover(df)
```

## Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(xtrain.shape)
print(xtest.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['location']),remainder='passthrough', force_int_remainder_cols=False )
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler, lr)
```

```python
pipe.fit(xtrain,ytrain)
```

```python
ypred = pipe.predict(xtest)
print(r2_score(ytest, ypred))
```

```python
lasso = Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(xtrain, ytrain)
```

```python
ypred_lasso = pipe.predict(xtest)
print(r2_score(ytest, ypred_lasso))
```

```python
ridge = Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(xtrain, ytrain)
```

```python
ypred_ridge= pipe.predict(xtest)
print(r2_score(ytest, ypred_ridge))
```
